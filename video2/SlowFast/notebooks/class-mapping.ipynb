### Obtain K600 labels in alphabetical order

```python
label_set_k_600 = set()
for k, v in train_kinetics_600.items():
    label_set_k_600.add(v['annotations']['label'])
    
label_set_k_600 = list(label_set_k_600)
```

```python
label_set_k_600_sorted = sorted(label_set_k_600)
label_set_k_600_sorted
```

```python
dict_k_600 = {i: e for i, e in enumerate(label_set_k_600_sorted)}
dict_k_600_rev = {v: k for k, v in dict_k_600.items()}
dict_k_600
```

## Form class mappings

```python
import spacy

nlp = spacy.load("en_core_web_md")
```

```python
dict_ava_classes, dict_ava_classes_rev
```

```python
dict_similarity = {}

for ava_c_name, ava_c_id in tqdm(dict_ava_classes.items()):
    
    dict_similarity[ava_c_id] = {}
    for k_c_id, k_c_name in enumerate(label_set_k_600_sorted):
        sim_score = nlp(ava_c_name).similarity(nlp(k_c_name))
        
        dict_similarity[ava_c_id][k_c_id] = sim_score
        
#         print(f"{ava_c_name} --> {k_c_name}: {sim_score}")
    
#     break
```

```python
for k, v in dict_similarity.items():
    
    top_10_K_classes = [
        (k, v) for k, v in sorted(v.items(), key=lambda item: item[1], reverse=True)
    ][:10]
    
    print(f"{dict_ava_classes_rev[k]} --> ")
    
    for e in top_10_K_classes[:3]:
        print(f"\t {dict_k_600[e[0]]}: {e[1]:.4f}")

# dict_similarity[0]
```
